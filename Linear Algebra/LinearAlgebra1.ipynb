Linear Algebra in Machine Learning

Why is Linear Algebra important for Machine Learning-
1. Foundation of ML and DL
2. Powers Neural Networks for faster and accurate modelling
3. Enables framework operations

Functional components of Linear Algebra - 
1. Scaler - a single digit number
2. Vector - 1D array
3. Matrix - 2D grid of row and columns
4. Tensor - 3D+ grid
